```python
import pandas as pd
import numpy as np
```

```python
data2023 = pd.read_excel("./dataset/2023.xlsx")
data2023
```

```python
data2019 = pd.read_excel('./dataset/2019.xlsx')
data2020 = pd.read_excel('./dataset/2020.xlsx')
data2021 = pd.read_excel('./dataset/2021.xlsx')
data2022 = pd.read_excel('./dataset/2022.xlsx')
```

```python
data = list()
data.append(data2019)
data.append(data2020)
data.append(data2021)
data.append(data2022)
data.append(data2023)
```

```python
months = list(data2019.columns)
months.remove('Date')
months
```

```python
data
```

```python
for item in data:
    for month in months:
        item[month].fillna(int(item[month].mean()), inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
#graph
linAQI = list()
for item in data:
    for month in months:
        linAQI.append(item[month].mean())
```

```python
len(linAQI)
```

```python
lineMonths = list()

for i in range (0,5):
    for j in range (0,len(months)):
        lineMonths.append(months[j])
```

```python
y = list()
for i in range (1,61):
    y.append(i)
y
```

```python
plt.figure(figsize=(15, 12))
plt.plot(y,linAQI)
plt.xlabel("Months")
plt.ylabel("AQI")
plt.show()
```

```python
health_data = pd.read_csv('./dataset/air_quality_health_impact_data.csv')
health_data.head()
```

```python
columns_to_drop = ['HealthImpactClass','WindSpeed','Humidity','Temperature']
```

```python
health_data = health_data.drop(columns=columns_to_drop,axis=1)
health_data.head()
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(health_data['PM10'],health_data['HealthImpactScore'] ,marker='.')
plt.xlabel("PM10 Particals")
plt.ylabel("Health Impact Score")
plt.show()
```

```python
health_data_final = health_data[health_data['HealthImpactScore'] != 100.00]
health_data_final.head()
```

```python
len(health_data_final)
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(health_data_final['PM10'],health_data_final['HealthImpactScore'] ,marker='.')
plt.xlabel("PM10 Particals")
plt.ylabel("Health Impact Score")
plt.show()
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(health_data_final['PM2_5'],health_data_final['HealthImpactScore'] ,marker='+',color='red')
plt.xlabel("PM10 Particals")
plt.ylabel("Health Impact Score")
plt.show()
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(health_data_final['NO2'],health_data_final['HealthImpactScore'] ,marker='^', color='green')
plt.xlabel("PM10 Particals")
plt.ylabel("Health Impact Score")
plt.show()
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(health_data_final['O3'],health_data_final['HealthImpactScore'] ,marker='*', color='brown')

plt.xlabel("PM10 Particals")
plt.ylabel("Health Impact Score")
plt.show()
```
